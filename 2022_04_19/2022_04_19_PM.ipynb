```python
#x,y 입력 받아서 어떤게 큰수 인지 어떤게 작은수 인지 비교 if input

x = int(input('첫번째 입력 하세요!: '))
y = int(input('두번째 입력 하세요!: '))

max_value = (x if x > y else y)
min_value = (y if x > y else x)

print('큰 수 =',max_value,'작은 수 =',min_value)
```
